Funções são trechos de código que você chamar para executar uma tarefa específica. Se no seu código algumas linhas são repetidos com frequência, você pode definir uma função e, ao invés de repetir aquele trecho ao longo do código, você pode chamar a função.


## 3.1. Funções sem retorno

Para criar uma função sem retorno, podemos escrever o seguinte trecho de código-fonte:



```python
def nomeDaFuncao(parametro_1,parametro_2, ..., parametro_n):

  '''
  corpo da função:
  Operações e comandos relacionados com a função
  '''

# parte geral do código-fonte (onde tudo começa!)

# chamada da função nomeDaFuncao
nomeDaFuncao(a_1,a_2,...,a_n)
```

em que:
  

*   **def** é uma palavra reservada do Python indicando que você estará criando uma função;
*   **nomeDaFuncao** é o nome que você deve definir para sua função;
*   **parametro_1,parametro_2, ..., parametro_n** são os parâmetros definidos, cada um separado por uma vírgula.
*   "*corpo da função*" é o trecho de código composto por operações e comandos que são executados quando a função for chamada
*   Na linha 11 ocorre a chamada da função e os argumentos que lhe são passados $a_1, a_2, \ldots, a_n$.

## 3.2. Escopo de variáveis em Python

Na linguagem Python, as variáveis podem possuir os seguintes escopos em relação à uma função:

- Escopo local;
- Escopo global;

### Escopo local

Uma variável é dita **local** se for criada e inicializada dentro de uma função pertence ao escopo local dessa função, sendo visível e podendo ser utilizada apenas dentro da própria função.

É o caso dos parâmetros $x$ e $y$, como também da variável $z$ na função **imprimeSoma**, que foram criados no escopo local da função. Se tentarmos imprimir o valor de $z$ na parte geral do código-fonte, receberemos uma mensagem de erro:


```python
# Código-fonte imprimeSoma_2.py

def imprimeSoma(x,y):
  z = x+y
  print(f'Valor da soma: {z}')

a = int(input("Digite um numero inteiro (a):"))
b = int(input("Digite outro numero inteiro (b):"))

imprimeSoma(a,b)

print(z)
```



### Nomes de Variáveis

Se você empregar o mesmo nome para duas variáveis, uma delas dentro e outra foram da função, o Python vai lidar com elas como duas variáveis separadas. Uma delas será associada ao escopo global enquanto que a outra estará ligada ao escopo local da função.

Repare no exemplo a seguir. A função **setNum** imprime a variável local $num$ e depois vai imprimir a variável global $x$:

```python
# Código-fonte setNum.py

def setNum(x):
  num = x
  print(num)

num = 15

setNum(3)

print(num)
```

No exemplo abaixo, temos a função divisaoResto, que recebe dois números inteiros $dividendo$ e $divisor$, calcula e retorna os valores do quociente e do resto:

```python
# Código-fonte divisaoResto.py

def divisaoResto(dividendo,divisor):
  quociente = dividendo // divisor
  resto = dividendo%divisor
  return quociente,resto

a = int(input("Digite um numero inteiro (a):"))
b = int(input("Digite outro numero inteiro (b):"))

q,r = divisaoResto(a,b)
print(f"Quociente: {q}; Resto: {r}")
```

Não é necessário pegar o retorno de todos os valores de uma função. Podemos utilizar o caractere "_" na posição do valor de retorno que queremos descartar (ou não utilizar). No exemplo abaixo, embora a função **divisaoResto** sempre retorna dois valores, estamos interessados apenas em pegar o valor do quociente:

```python
# Código-fonte divisaoResto.py

def divisaoResto(dividendo,divisor):
  quociente = dividendo // divisor
  resto = dividendo%divisor
  return quociente,resto

a = int(input("Digite um numero inteiro (a):"))
b = int(input("Digite outro numero inteiro (b):"))

q,_ = divisaoResto(a,b)
print(f"Quociente: {q}")
```

Vale ressaltar que o caractere "_" na linha 11 não é uma variável, mas sim um mecanismo para não guardar o segundo valor retornado.
